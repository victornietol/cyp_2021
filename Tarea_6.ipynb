```
#Problema 3.3 

SERIE = 0

N = int(input("Ingresa el número: "))
BAND = True
I = 1

while I <= N:
  if BAND == True:
    SERIE += 1/I
    BAND = False
  else:
    SERIE -= 1/I
    BAND = True
  I += 1
print(f"\nEl resultado de la serie es {round(SERIE,4)}")    
```

```
#Problema 3.5

N = 0
NUM = 0
SUMOTR = 0 
SUMPOS = 0
CUEPOS = 0

N = int(input("Ingresa el número de datos: "))
I = 1
while I <= N:
  NUM = int(input(f"Ingresa el valor del dato {I}: "))
  if NUM > 0:
    SUMPOS += NUM
    CUEPOS += 1
  else:
    SUMOTR += NUM
  I += 1
PROGEN = (SUMPOS + SUMOTR) / N
PROPOS = (SUMPOS / CUEPOS)
print(f"\nLa cantidad de números mayores que cero fueron {CUEPOS}\nEl promedio de los números positivos es {PROPOS}")
print(f"El promedio en general de los números es {PROGEN}")
```

```
#Problema 3.15

DUR = 0
CL = 0
CUENTA = 0.0
DUR = 0
COSTO = 0.0
CONT = 1

print("Si la llamada es de tipo LOCAL escriba        : L\nSi la llamada es de tipo NACIONAL escriba     : N\nSi la llamada es de tipo INTERNACIONAL escriba: I")
TIPO = input(f"\nIngrese el tipo de la llamada {CONT}: ")
TIPO = TIPO.upper()
DUR = int(input("Ingrese la duración de la llamada: "))
while TIPO != "X" and DUR != -1:
  if TIPO == "I":
    if DUR > 3:
      COSTO = 7.59 + ((DUR - 3) * 3.03)
    else:
      COSTO = 7.59
  if TIPO == "L":
    CL += 1
    if CL > 50:
      COSTO = 0.60
    else:
      COSTO = 0
  if TIPO == "N":
    if DUR > 3:
      COSTO = 1.20 + ((DUR - 3) * 0.48)
    else:
      COSTO = 1.20
  CUENTA += COSTO
  TIPO = input(f"\nIngrese el tipo de la llamada {CONT}: ")
  TIPO = TIPO.upper()
  DUR = int(input(f"Ingrese la duración de la llamada: ")) 
  CONT += 1
CONT -= 2 
print(f"\nLa cuenta total a pagar por las {CONT} llamadas es de: ${round(CUENTA,2)}")
```

```
#Problema 3.17

I = 3
SP = 0
M = 0
J = 0

M = int(input("Introduce un número: "))
if M >= 1:
  SP += 1 
  print("\nNúmero primo:",1)
  if M >= 2:
    SP += 1
    print("Número primo:",2)
while I <= M:
  BAND = True
  J = 3
  while J < (I / 2) and BAND == True:
    if (I % J) == 0:
      BAND = False
    J += 2
  if BAND == True:
    print(f"Número primo: {I}")
    SP += 1
  I += 2
print(f"\nEntre 1 y {M}, hay {SP} números primos.")
```

```
#Problema 3.18

NFAB = 0
FAB = 0
MES = 0.0
MAYPRO = 0

NFAB = int(input("Ingrese el número de fábricas: "))
if NFAB <= 100:
  I = 1
  while I <= NFAB:
    FAB = int(input(f"\nIngrese la clave de la fábrica {I}: "))
    TOTANU = 0
    J = 1
    while J <= 12:
      MES = float(input(f"Ingrese la producción de la fábrica {FAB} en el mes {J}: $"))
      TOTANU += MES
      if J == 7 and  MES > 3000000:
        print(f"Clave de la fábrica: {FAB}")
      J += 1
      if TOTANU > MAYPRO:
        MAYPRO = TOTANU
        CLAVE = FAB
    print(f"\nLa producción anual de la fábrica {FAB} es: ${TOTANU}")
    I += 1
  print(f"\nClave de la fábrica con mayor produccón: {CLAVE}\nCon una producción de: ${MAYPRO}")
else:
  print("Número inválido de fábricas.")
```
