```
COLECCION = {"Mario": 20, "Mariana": 19, "Yessica": 14, "Luis": 18}

for i in COLECCION: 
  print(f"{i} -> {COLECCION[i]}")
```

```
COLECCION = {"Mario": 20, "Mariana": 19, "Yessica": 14, "Luis": 18}

for CLAVE, VALOR in COLECCION.items():
  print(f"{CLAVE} -> {VALOR}")
```

```
EDAD = int(input("Introduce tu edad: "))

while EDAD < 0:
  print("\nHas introducido una edad negativa. Intententalo de nuevo")
  EDAD = int(input("\nIntroduce tu edad: "))
print("\nGracias por participar. Puede pasar")
print("Edad del aspirante: " + str(EDAD))
```

```
EDAD = int(input("Introduce tu edad: "))

while EDAD < 0 or EDAD > 100:
  print("\nHas introducido una edad incorrecta. Intententalo de nuevo")
  EDAD = int(input("\nIntroduce tu edad: "))
print("\nGracias por participar. Puede pasar")
print("Edad del aspirante: " + str(EDAD))
```
