# Slicing (continuación)

[ inicio : stop : incremento ]

Notas: 
1. El stop no es incluido.
1. El incremento se puede modificar.
1. Soporta indexado alternativo (negativos)

```
FRASE = "Como hacer a la gente mas positiva?, muy fácil, has que pierdan un electron¡"
print(FRASE)
  # Imprimir "ace" de la palabra hacer 
print(FRASE[6:9:1])


# Valores por defecto

print(FRASE[ : 5 : ])    # Incremento = 1


print(len(FRASE))  # Longitud de frase 
# Incremento
print(FRASE[ ::2]) # Toda la frase
print(FRASE[-1::-1])  # Inverso   -77

NOMBRE = "Juan José"
print(NOMBRE[1::-1])

PRIMER = NOMBRE[:4:]
print(PRIMER)

print(NOMBRE[-9::1])
```

# Métodos de la clase String(str)
En python todo es un objeto (P.O.O.), eso significa que cada tipo de dato trae consigo métodos (funciones) incluidas.

Para identificar los metodos disponibles se puede usar una de dos funciones:
1. dir(str) # Resumen de los metodos de esa clase.
1. help(str) # Información completa de los metodoos de esa clase. 

```
NOMBRE = "Juan José"
print(NOMBRE)
print(dir(str))
```

Los nombre de métodos que inician con doble guion bajo, significa que son métodos privados (POO), y no tendrías por que usarlos como programador consumidor.
Los demás métodos son públicos y son de libre uso.

```
print(NOMBRE)
print(NOMBRE.islower())
NOMBRE = "PEDRO PICAPIEDRA"
print(NOMBRE.capitalize())
print(NOMBRE.endswith("DRA"))
print(NOMBRE.replace("E","***"))

# Para una sola letra
ind_e = NOMBRE.find("E")
print(ind_e)
NOMBRE = NOMBRE[:ind_e:]+ "@" + NOMBRE[ind_e+1::]
print(NOMBRE)
print(NOMBRE.replace("E","@",0))   # El 0 es el lugar del valor a remplazar

print(NOMBRE.upper())
NOMBRE = "josé josé, el principe de la canción"
print(NOMBRE.count("josé"))
```

```
# Metodos para pocesar Strings

NUMEROS = " 2 ,  1,3,4,     5,45 ,  7     "
print(NUMEROS)

# Como extraigo los numeros de la cadena
# 1.- Eliminar los espacios al principio y al final con strip()
NUMEROS = NUMEROS.strip()
print(NUMEROS)

# 2.- Separarlo por comas 
# Dividir en segmentos una cadena con  .split()    ,  (Genera una lista)
NUMEROS = NUMEROS.split(",")
print(NUMEROS)

# Paso 3, con un fo re indices limpiar cadenas de lista
for indice in range(len(NUMEROS)):
  NUMEROS[indice] = NUMEROS[indice].strip()
print(NUMEROS)

# Paso 4, convertirlos a tipo entero
for indice in range(len(NUMEROS)):
  NUMEROS[indice] = int(NUMEROS[indice])
print(NUMEROS)

# Paso 5, usar los números
```
